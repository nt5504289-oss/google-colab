```python
 import numpy as np
 import matplotlib.pyplot as plt
```

```python
#line plot
xpoint = np.array([1,3])
ypoint = np.array([2,4])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
#line plot
xpoint= np.array([5,8])
ypoint = np.array([6,9])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
#line plot
student=np.array([2,4,6,7,9,2])
passed = np.array([1,2,3,4,5,6])
font1={'family':'serif','color':"blue",'size':15}
font2={'family':'serif','color':"red",'size':20}
plt.plot(student,passed,"*:b",ms='13',mfc="blue",mec="orange")
plt.xlabel("student",fontdict=font1)
plt.ylabel("passed",fontdict=font1)
plt.title("student vs passed",fontdict=font2)
plt.show()

```

```python
cgpa = np.array([6,6.5,7,7.8,8.6,8,9,9.2,9.5,10])
package = np.array([4,4.5,5,5.8,6.2,6.8,7,7.5,8,8.8])
font1= {'family':'serif','color':'blue','size':15}
font2= {'family':'serif','color':'darked','size':20}
plt.plot(cgpa,package,"*:b", ms=15,mec="orange",mfc="red",linewidth=2.4)
plt.xlabel("CGPA",fontdict=font1)
plt.ylabel("PACKAGE", fontdict = font1)
plt.title("CGPA VS PACKAGE",fontdict= font2)
plt.show()
```

```python
#bar plot--it is work on numerical as well as categorical data
fruit = np.array("Apple","Banana","Orange","Watermelon")
quantity=np.array([80,120,50,100])
color = np.array(["red","yellow","orange","green"])
plt.bar(fruit,quantity,color = color)
plt.show()
```

```python
import numpy as np
from numpy.linalg import norm
A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])
# compute cosine similarity
cosine = np.dot(A,B) / (norm(A)* norm(B))
print("Cosine Similarity:",cosine)
```
